```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()

X_train_tfidf = vectorizer.fit_transform(X_train)
X_train_tfidf.shape
```

```python
from sklearn.svm import LinearSVC
clf = LinearSVC()
clf.fit(X_train_tfidf,y_train)
```

```python
from sklearn.pipeline import Pipeline
text_clf = Pipeline([('tfidf', TfidfVectorizer()),
                     ('clf', LinearSVC()),])


text_clf.fit(X_train, y_train)  


predictions = text_clf.predict(X_test)
```

```python
from sklearn import metrics
print(metrics.confusion_matrix(y_test,predictions))
```

```python
print(metrics.classification_report(y_test,predictions))
```

```python
print(metrics.accuracy_score(y_test,predictions))
```

```python
import pandas as pd

# Column names as provided
column_names = [
    "Disease", 
    "Symptom_1", "Symptom_2", "Symptom_3", "Symptom_4", 
    "Symptom_5", "Symptom_6", "Symptom_7", "Symptom_8", 
    "Symptom_9", "Symptom_10", "Symptom_11", "Symptom_12", 
    "Symptom_13", "Symptom_14", "Symptom_15", "Symptom_16", 
    "Symptom_17"
]

# Step 1: Load the dataset
df = pd.read_csv('corrected_dataset.csv', names=column_names)

# Step 2: Initialize empty DataFrame for symptoms
symptoms_df = pd.DataFrame(index=df.index)

# Step 3: Process each row to create symptom columns
unique_symptoms = set()
for i in range(1, 18):
    unique_symptoms.update(df[f'Symptom_{i}'].dropna().unique())

# Remove any empty strings or invalid values
unique_symptoms = {symptom for symptom in unique_symptoms if isinstance(symptom, str) and symptom.strip()}

# Create a column for each symptom
for symptom in unique_symptoms:
    # Initialize the column with zeros
    symptoms_df[symptom] = 0
    # Check each symptom column for this symptom
    for i in range(1, 18):
        symptoms_df[symptom] = symptoms_df[symptom] | (df[f'Symptom_{i}'] == symptom)

# Convert to int type
symptoms_df = symptoms_df.astype(int)

# Keep the original Disease column
symptoms_df['Disease'] = df['Disease']

# Reorder columns to have Disease first
cols = ['Disease'] + [col for col in symptoms_df.columns if col != 'Disease']
symptoms_df = symptoms_df[cols]

# Print some information
print("Shape of encoded DataFrame:", symptoms_df.shape)
print("\nColumns in the DataFrame:")
print(symptoms_df.columns.tolist())
print("\nFirst few rows:")
print(symptoms_df.head())

# Save the encoded DataFrame
symptoms_df.to_csv('training.csv', index=False)

# Print number of unique symptoms
print(f"\nNumber of unique symptoms: {len(unique_symptoms)}")
```
