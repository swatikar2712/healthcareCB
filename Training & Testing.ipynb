```python
import pandas as pd
from sklearn.model_selection import train_test_split


column_names = [
    "Disease", 
    "Symptom_1", "Symptom_2", "Symptom_3", "Symptom_4", 
    "Symptom_5", "Symptom_6", "Symptom_7", "Symptom_8", 
    "Symptom_9", "Symptom_10", "Symptom_11", "Symptom_12", 
    "Symptom_13", "Symptom_14", "Symptom_15", "Symptom_16", 
    "Symptom_17"
]


df = pd.read_csv('dataset.csv', names=column_names)


unique_symptoms = set()
for col in [f"Symptom_{i}" for i in range(1, 18)]:
    unique_symptoms.update(df[col].dropna().unique())


unique_symptoms = sorted({symptom for symptom in unique_symptoms 
                        if isinstance(symptom, str) and symptom.strip()})


encoded_df = pd.DataFrame()
encoded_df['Disease'] = df['Disease']


for symptom in unique_symptoms:
   
    encoded_df[symptom] = 0

    for i in range(1, 18):
       
        mask = df[f'Symptom_{i}'] == symptom
        encoded_df.loc[mask, symptom] = 1


train_df, test_df = train_test_split(
    encoded_df,
    test_size=0.33,
    random_state=42
)


train_df.to_csv('training.csv', index=False)
test_df.to_csv('testing.csv', index=False)


print("Data has been split and encoded.")
print(f"Training set shape: {train_df.shape}")
print(f"Testing set shape: {test_df.shape}")


print("\nColumns in the datasets:")
print("First 10 columns:", list(train_df.columns[:10]))


print("\nSample of training data (first 5 rows, first 10 columns):")
print(train_df.iloc[:5, :10])


intermediate_cols = [col for col in train_df.columns if 'Symptom_' in col]
print("\nChecking for intermediate symptom columns:", 
      "None found" if not intermediate_cols else "Found: " + str(intermediate_cols))


sample_symptom = list(train_df.columns)[1] 
print(f"\nValue counts for {sample_symptom}:")
print(train_df[sample_symptom].value_counts())
```
